```python
import torch
from diffusers import StableDiffusionPipeline
from PIL import Image  # For resizing the image
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Install necessary libraries
!pip install diffusers transformers accelerate scipy ftfy pillow

# Load the Stable Diffusion pipeline
pipe = StableDiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-2-1-base", torch_dtype=torch.float32)
pipe = pipe.to("cpu")

# Define the prompt and generate the image
prompt = "elephant flies in sky"
image = pipe(prompt, num_inference_steps=20).images[0]

# Resize the image
new_size = (1024, 1024)  # Specify the desired size (e.g., 1024x1024)
resized_image = image.resize(new_size)  # No need for Image.ANTIALIAS

# Save the resized image
resized_image.save("resized_generated_image.png")

# Display the resized image
img = mpimg.imread("resized_generated_image.png")
plt.imshow(img)
plt.axis('off')  # Optional: Turn off axis labels
plt.show()
```
